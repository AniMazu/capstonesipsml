```python
def LowerUnderscoreUpper(s):
    l = len(s)

    for i in range(0,l-2):
        if s[i].islower() and s[i+1] == '_' and s[i+2].isupper():
            return True
    return False

def HasUnderscore(s):
    s = str(s)

    for c in s:
        if c == '_':
            return True
    return False

def LowerUpperLower(s):
    s = str(s)

    for i in range(0, len(s)-2):
        if s[i].islower() and s[i+1].isupper() and s[i+2].islower():
            return True
    return False

def MultipleLowerUpperLower(s):
    s = str(s)

    flag = False
    for i in range(0, len(s)-2):
        if s[i].islower() and s[i+1].isupper() and s[i+2].islower():
            if flag:
                return True
            else:
                flag = True
    return False

def ExactlyTwoUppercase(s):
    count = 0

    for c in s:
        if c.isupper():
            count+=1
    if count ==2:
        return True
    else:
        return False

def AllLowerMoreThan(bound, s):
    s = str(s)

    if len(s) < bound:
        return False
    for c in s:
        if c.isupper():
            return False
    return True

def AdjacentUppers(s):
    s = str(s)

    for i in range(0,len(s)-1):
        if s[i].isupper() and s[i+1].isupper():
            return True
    return False

def StartLetterEndNonLetter(s):
    s = str(s)

    if s[0].isalpha() and not s[-1].isalpha():
        return True
    return False


# Hostname specific features
def LengthGTLT(gt, ls, s):
    l = len(str(s))
    if l >= gt and l <= ls:
        return True
    else:
        return False

def IllegalHostnameChars(s):
    s = str(s)
    illegal_chars = [".", "\", ""/", "*", "?", "\"", "<", ">", "|", ",", "~", ":", "!", "@", "#", "$", "%", "^", "&", "'", "(", ")", "{", "}", " "]
    for c in s:
        if c in illegal_chars:
            return True
    return False

def AlphaOrDigit(s):
    s = str(s)
    if s.isalpha() or s.isdigit():
        return True
    return False

def HostIllegalEnding(s):
    s = str(s)
    if s[-1] == '-' or s[-1] == '.':
        return True
    return False

# Filename specific features

def ContainsPeriod(s):
    s = str(s)
    for c in s:
        if c == '.':
            return True
    return False

def IsUrl(s):
    s = str(s)
    l = len(s)
    if l > 3:
        if s[0:4] == "http":
            return True
    if l > 4:
        if s[0:5] == "https":
            return True
    return False

def HasSlash(s):
    s = str(s)
    for c in s:
        if c == '/' or c == '\\':
            return True
    return False

def HasMultipleSlash(s):
    s = str(s)
    slashes = ['/','\\']
    flag = False
    for c in s:
        if not flag and c in slashes:
            flag = True
        elif flag and c in slashes:
            return True
    return False

def HasPossibleExtension(s):
    s = str(s)
    for i in range(0,len(s)):
        if s[i] == '.':
            x = s[i:-1]
            if len(x) >= 2 and len(x) <= 4:
                return True
    return False

# IP Features
def NumbersThenPeriod(s):
    s = str(s)
    freq = 0
    for i in range(0,len(s)-1):
        if s[i] >= '0' and s[i] <= '9' and s[i+1] == '.':
            freq += 1
    return True if freq == 3 else False

def AtLeastFourDigits(s):
    s = str(s)
    counter = 0
    for c in s:
        if c.isnumeric():
            counter += 1
    return True if counter >=4 else False

# Servername Features

def HasPeriodAndSlash(s):
    s = str(s)
    period = False
    slash = False
    for c in s:
        if c == '.':
            period = True
        if s =='/':
            slash = True
    return True if period and slash else False

def HasInstanceNumPeriod(s):
    s = str(s)
    for i in range(0,len(s)-1):
        if s[i].isnumeric() and s[i+1] == '.':
            return True
    return False

def HasMultipleNumPeriod(s):
    s = str(s)
    flag = False
    for i in range(0,len(s)-1):
        if s[i].isnumeric() and s[i+1] == '.' and not flag:
            flag = True
        if s[i].isnumeric() and s[i+1] == '.' and flag:
            return True
    return False
```

```python
# Test Methods

class TestDriver(unittest.TestCase):

    # def testEndInANumber(self):
    #     self.assertEqual(EndInANumber("hello9"), True)
    #     self.assertEqual(EndInANumber("hello99o"), False)
    #     self.assertEqual(EndInANumber("12902"), True)

    # def testContainsSpecialChar(self):
    #     self.assertEqual(ContainsSpecialChar("!hello"), True)
    #     self.assertEqualContainsSpecialChar(("hello!"), True)
    #     self.assertEqual(ContainsSpecialChar("hello"), False)

    # def testNumberOfCapitalLetters(self):
    #     self.assertEqual(NumberOfCapitalLetters("HHello"), 2)
    #     self.assertEqual(NumberOfCapitalLetters("HHelloO"), 3)
    #     self.assertEqual(NumberOfCapitalLetters("hello"), 0)

    # def testLengthOfString(self):
    #     self.assertEqual(LengthOfString("hello"), 5)
    #     self.assertEqual(LengthOfString("!hello"), 6)
    #     self.assertEqual(LengthOfString("!hello&O"), 8)

    # def testFeaturesJsonSearch(self):
    #     self.assertEqual(FeaturesJsonSearch(""), True)
    #     self.assertEqual(FeaturesJsonSearch(""), True)
    #     self.assertEqual(FeaturesJsonSearch(""), True)

    # def testCalcAccuracy(self):
    #     self.assertEqual((CalcAccuracy), True)
    #     self.assertEqual((CalcAccuracy), True)
    #     self.assertEqual((CalcAccuracy), True)

    def testLowerUnderscoreUpper(self):
        self.assertEqual(LowerUnderscoreUpper("hello_World"), True)
        self.assertEqual(LowerUnderscoreUpper("hello_world"), False)
        self.assertEqual(LowerUnderscoreUpper("helloWorld"), False)

    def testHasUnderscore(self):
        self.assertEqual(HasUnderscore("hello_World"), True)
        self.assertEqual(HasUnderscore("hello&_&World"), True)
        self.assertEqual(HasUnderscore("helloWorld"), False)

    def testLowerUpperLower(self):
        self.assertEqual(LowerUpperLower("hellOworld"), True)
        self.assertEqual(LowerUpperLower("helLOworld"), False)
        self.assertEqual(LowerUpperLower("hellOWorld"), False)

    def testMultipleLowerUpperLower(self):
        self.assertEqual(MultipleLowerUpperLower("hElLo"), True)
        self.assertEqual(MultipleLowerUpperLower("hEllo"), False)
        self.assertEqual(MultipleLowerUpperLower("hELLO"), False)

    def testExactlyTwoUppercase(self):
        self.assertEqual(ExactlyTwoUppercase("HellO"), True)
        self.assertEqual(ExactlyTwoUppercase("hELlO"), False)
        self.assertEqual(ExactlyTwoUppercase("hELLO"), False)

    def testAllLowerMoreThan(self):
        self.assertEqual(AllLowerMoreThan(3, "hello"), True)
        self.assertEqual(AllLowerMoreThan(3, "helLo"), False)
        self.assertEqual(AllLowerMoreThan(10, "hello"), False)

    def testAdjacentUppers(self):
        self.assertEqual(AdjacentUppers("HEllo"), True)
        self.assertEqual(AdjacentUppers("HeLlO"), False)
        self.assertEqual(AdjacentUppers("hello"), False)

    def testStartLetterEndNonLetter(self):
        self.assertEqual(StartLetterEndNonLetter("Hello!"), True)
        self.assertEqual(StartLetterEndNonLetter("!Hello!"), False)
        self.assertEqual(StartLetterEndNonLetter("Hello"), False)
        self.assertEqual(StartLetterEndNonLetter("!Hello"), False)

    # Hostname specific features
    def testLengthGTLT(self):
        self.assertEqual(LengthGTLT(0, 10, "hello"), True)
        self.assertEqual(LengthGTLT(0, 2, "hello"), False)
        self.assertEqual(LengthGTLT(0, 5, "hello"), True)

    def testIllegalHostnameChars(self):
        self.assertEqual(IllegalHostnameChars("Hello*"), True)
        self.assertEqual(IllegalHostnameChars("Hello"), False)
        self.assertEqual(IllegalHostnameChars("^&Hello"), True)

    def testAlphaOrDigit(self):
        self.assertEqual(AlphaOrDigit("Hello*"), False)
        self.assertEqual(AlphaOrDigit("Hello"), True)
        self.assertEqual(AlphaOrDigit("^&Hello"), False)
        self.assertEqual(AlphaOrDigit("73110"), True)

    def testHostIllegalEnding(self):
        self.assertEqual(HostIllegalEnding("Hello-"), True)
        self.assertEqual(HostIllegalEnding("Hello"), False)
        self.assertEqual(HostIllegalEnding("Hel_lo"), False)
        self.assertEqual(HostIllegalEnding("73110!"), False)

    # Filename specific features

    def testContainsPeriod(self):
        self.assertEqual(ContainsPeriod("Hel.lo"), True)
        self.assertEqual(ContainsPeriod("Hello."), True)
        self.assertEqual(ContainsPeriod(".Hello"), True)
        self.assertEqual(ContainsPeriod("Hello"), False)

    def testIsUrl(self):
        self.assertEqual(IsUrl("https://www.microsoft.com/en-us/"), True)
        self.assertEqual(IsUrl("htt.ps://www.microsoft.com/en-us/"), False)
        self.assertEqual(IsUrl("https|||://www.microsoft.com/en-us/"), True)

    def testHasSlash(self):
        self.assertEqual(HasSlash("hello/world"), True)
        self.assertEqual(HasSlash("hello\world"), True)
        self.assertEqual(HasSlash("hello//world"), True)
        self.assertEqual(HasSlash("hello\\world"), True)
        self.assertEqual(HasSlash("helloWorld"), False)

    def testHasMultipleSlash(self):
        self.assertEqual(HasMultipleSlash("hello//World"), True)
        self.assertEqual(HasMultipleSlash("he/llo/World//"), True)
        self.assertEqual(HasMultipleSlash("hel/lo/World"), True)
        self.assertEqual(HasMultipleSlash("he\llo/World\\"), True)
        self.assertEqual(HasMultipleSlash("hel\lo\World"), True)
        self.assertEqual(HasMultipleSlash("helloWorld"), False)

    def testHasPossibleExtension(self):
        self.assertEqual(HasPossibleExtension("Hello.mp4"), True)
        self.assertEqual(HasPossibleExtension("Hello.png"), True)
        self.assertEqual(HasPossibleExtension("notavirus.virus"), False)
        self.assertEqual(HasPossibleExtension("hello."), False)
        self.assertEqual(HasPossibleExtension("hello"), False)

    # IP Features
    def testNumbersThenPeriod(self):
        self.assertEqual(NumbersThenPeriod("132.1.13.102"), True)
        self.assertEqual(NumbersThenPeriod("13.12.1.12ello"), True)     # Fails
        self.assertEqual(NumbersThenPeriod("13_12_102_203"), False)
        self.assertEqual(NumbersThenPeriod("hello1.12.13h.15llo"), False)

    def testAtLeastFourDigits(self):
        self.assertEqual(AtLeastFourDigits("4444hello"), True)
        self.assertEqual(AtLeastFourDigits("hello4444"), True)
        self.assertEqual(AtLeastFourDigits("4h4el4lo"), False)
        self.assertEqual(AtLeastFourDigits("4h4el4lo4"), True)
        self.assertEqual(AtLeastFourDigits("hello4"), False)

    # Servername Features
    def testHasPeriodAndSlash(self):
        self.assertEqual(HasPeriodAndSlash("./hello"), True)    # Fails
        self.assertEqual(HasPeriodAndSlash("./helloWorld./"), True)
        self.assertEqual(HasPeriodAndSlash("hello\."), False)
        self.assertEqual(HasPeriodAndSlash("hello./"), True)
        self.assertEqual(HasPeriodAndSlash(".hello/"), True)

    def testHasInstanceNumPeriod(self):
        self.assertEqual(HasInstanceNumPeriod("hello9."), True)
        self.assertEqual(HasInstanceNumPeriod("hello9"), False)
        self.assertEqual(HasInstanceNumPeriod("hello."), False)
        self.assertEqual(HasInstanceNumPeriod("9hello."), False)

    def testHasMultipleNumPeriod(self):
        self.assertEqual(HasMultipleNumPeriod("...hello."), False)
        self.assertEqual(HasMultipleNumPeriod(".hello."), False)
        self.assertEqual(HasMultipleNumPeriod("hello."), False)
        self.assertEqual(HasMultipleNumPeriod(".555.hello"), True)
```
